```python
# Import the pandas library, which is used for data manipulation and analysis
import pandas as pd

# This ID comes from the spreadsheet URL and uniquely identifies the file.
SPREADSHEET_ID = '1OGs8KoHGEZDxyFdaxAKmZbeO1BiSWVQ6'
# Define the Sheet GID (Grid ID) which identifies a specific sheet (tab) within the spreadsheet.
SHEET_GID = '361756276' 

# This allows pandas to read the sheet directly from the web.
url = f'https://docs.google.com/spreadsheets/d/{SPREADSHEET_ID}/export?format=csv&gid={SHEET_GID}'
```

```python
# Read the CSV data from the Google Sheet into a pandas DataFrame.
df = pd.read_csv(url)
# Display the DataFrame to visually inspect the imported data.
df
```

```python
# This is useful when the spreadsheet contains metadata or notes above the actual header row.
df = pd.read_csv(url, header=20)
# Display the DataFrame after reloading it with the correct header row.
df
```

```python
# Rename selected columns to more meaningful and readable names.
df = df.rename(columns={"OdName": "Country","AreaName": "Continent","RegName": "Region" })
df
```

```python
# These columns will be removed from the dataset.
columns_to_drop = ['Type', 'AREA','Coverage','REG','DEV']
```

```python
# Drop the unnecessary columns from the DataFrame.
# The resulting DataFrame contains only the relevant columns.
df_remaining = df.drop(columns=columns_to_drop)
# Display the cleaned DataFrame with the selected columns retained.
df_remaining
```
